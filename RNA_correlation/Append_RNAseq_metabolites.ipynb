```python
import pandas as pd
import networkx as nx
import numpy as np
from scipy.stats import zscore
```

```python
rnaseq_data = pd.read_csv("D://CCLE_data/CCLE/CCLE_RNAseq_rsem_genes_tpm_20180929.txt/CCLE_RNAseq_rsem_genes_tpm_20180929.txt", sep = "\t")
```

```python
rnaseq_columns = []
for item in rnaseq_data.columns:
    item = item.split("_")[0]
    rnaseq_columns.append(item)
    
rnaseq_data.columns = rnaseq_columns
rnaseq_data = rnaseq_data.set_index("gene")
```

```python
import mygene
```

```python
geneids = rnaseq_data.index.tolist()
geneids = [i.split(".")[0] for i in geneids]
geneids_test = geneids[:20]
rnaseq_data.index = geneids
```

```python
mg = mygene.MyGeneInfo()
gene_symbols = mg.querymany(geneids, scopes='ensembl.gene', species = "human", fields = "symbol", returnall = False, as_dataframe = True)
```

```python
gene_symbols_found = gene_symbols[gene_symbols["notfound"] != True]
gene_symbols_found = gene_symbols_found[["_id", "symbol"]]
```

```python
combined_data = rnaseq_data.merge(gene_symbols_found, left_index = True, right_index = True)
combined_data = combined_data.drop("_id", axis =1)
combined_data = combined_data.drop("transcript", axis =1)
combined_data = combined_data.drop_duplicates(subset = "symbol")
combined_data =combined_data.set_index("symbol")
```

```python
combined_data
```

```python
celllines_mapped = pd.read_csv("../MCF7_removal/Celllines_mapping_manual_noMCF7.csv")
# Get only cell lines with ID
celllines_mapped = celllines_mapped.drop_duplicates(subset = ["dsIdx"])
celllines_mapped = celllines_mapped.dropna(subset = ["ID"])
sample_data_tissue = celllines_mapped["DI"].tolist()

celllines_replaced = []
for item in sample_data_tissue:
    item = item.replace("-", "")
    item = item.replace(".", "")
    item = item.replace(" ", "")
    item = item.split("[")[0]
    item = item.upper()
    celllines_replaced.append(item)
```

```python
metabolomics_data = pd.read_csv("../MCF7_removal/Metabolomics_data_noMCF7.csv", index_col = "ionIdx")
```

```python
metabolomics_data = metabolomics_data.drop("ionMz", axis =1)
```

```python
average_celllines = []

for item in celllines_mapped["ID"].unique():
    ids = celllines_mapped[celllines_mapped["ID"] == item]["dsIdx"].tolist()
    ids = [str(i) for i in ids] 
    metabolomics_data_subset = metabolomics_data[ids]
    metabolomics_data_subset[item] = metabolomics_data_subset.mean(axis =1)
    average_df = metabolomics_data_subset[item]
    average_celllines.append(average_df)
```

```python
averaged_metabolites = pd.concat(average_celllines, axis =1)
```

```python
updated_columns = []
for item in averaged_metabolites.columns:
    item = item.replace("-", "")
    item = item.replace(".", "")
    item = item.replace(" ", "")
    item = item.split("[")[0]
    item = item.upper()
    updated_columns.append(item)
    
averaged_metabolites.columns = updated_columns
```

```python
averaged_metabolites
```

```python
metabolites_rnaseq_combined = pd.concat([averaged_metabolites, combined_data], join = "inner")
```
