```python
import pandas as pd
```

```python
metabolites = pd.read_csv("./Metabolite_reference_table.csv")
```

```python
metabolites
```

```python
KEGG_ids = pd.read_csv("./name_map.csv")
```

```python
KEGG_ids = KEGG_ids[["Query", "KEGG"]]
```

```python
updated_table = metabolites.merge(KEGG_ids, left_on = "id", right_on = "Query")
```

```python
updated_table = updated_table.drop("id", axis =1)
```

```python
updated_table.to_csv("KEGG_id_table.csv")
```

```python
cellline_ids = pd.read_csv("./sample_info.csv")
```

```python
cellline_table = pd.read_csv("./cellline_table.csv")
```

```python
cellline_table.head()
```

```python
cellline_ids.head()
```

```python
cellline_subset = cellline_ids[["stripped_cell_line_name", "RRID"]]
```
