```python
gene = "MDM2"
MDM2_signalling = pd.read_csv("./MDM2_signalling/pathway_results.csv")
MDM2_signalling.columns = [gene + "Pathway", "Total", "Expected", "Hits", "pbase",  gene + "log10pbase", "Holm adjust", "FDR", gene + "Impact"]

MDM2_signalling_essential = MDM2_signalling[[gene + "Pathway", gene + "log10pbase", gene+ "Impact"]]

merged = TP53_signalling_essential.merge(MDM2_signalling_essential, left_on = "Pathway", right_on= gene +"Pathway")

merged["pvalueDifference"] = merged[gene + "log10pbase"] - merged["log10pbase"]
```
